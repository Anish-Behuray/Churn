```python
import numpy as np
import pandas as pd
import re
from ast import literal_eval

# Step 1: Clean existing rate column
def clean_rating(val):
    if isinstance(val, str) and '/' in val:
        return float(val.split('/')[0])
    try:
        return float(val)
    except:
        return np.nan

df['rate'] = df['rate'].apply(clean_rating)

# Step 2: For rows where 'rate' is NaN, extract from 'reviews_list'
def extract_rating_from_reviews(review_text):
    try:
        # Convert string to list of tuples
        review_list = literal_eval(review_text)
        ratings = []
        for entry in review_list:
            match = re.search(r'Rated\s+(\d+(\.\d+)?)', entry[0])
            if match:
                ratings.append(float(match.group(1)))
        if ratings:
            return round(np.mean(ratings), 1)  # Take average
    except:
        pass
    return np.nan

# Apply only to missing rates
df.loc[df['rate'].isnull(), 'rate'] = df.loc[df['rate'].isnull(), 'reviews_list'].apply(extract_rating_from_reviews)
```

```python
import numpy as np

# Define a function to extract second-last part from address
def extract_second_last(address):
    try:
        parts = [p.strip() for p in str(address).split(',')]
        if len(parts) >= 2:
            return parts[-2]
        else:
            return np.nan
    except:
        return np.nan

# Apply only where location is missing or empty
df['location'] = df.apply(
    lambda row: extract_second_last(row['address']) if pd.isnull(row['location']) or str(row['location']).strip() == '' else row['location'],
    axis=1
)
```

Frustration index

```python
import ast

# Combine all review texts into one string per row
def extract_reviews(review_str):
    try:
        reviews = ast.literal_eval(review_str)
        return ' '.join([r[1] for r in reviews if isinstance(r, tuple) and len(r) > 1])
    except:
        return ''

df['combined_reviews'] = df['reviews_list'].apply(extract_reviews)
```

```python
from textblob import TextBlob

def get_sentiment(text):
    return TextBlob(text).sentiment.polarity  # Range: -1 (negative) to 1 (positive)

df['sentiment_score'] = df['combined_reviews'].apply(get_sentiment)
```

```python
# Normalize rate and sentiment to 0–1
df['norm_rate'] = (df['rate'] - df['rate'].min()) / (df['rate'].max() - df['rate'].min())
df['norm_sentiment'] = (df['sentiment_score'] - df['sentiment_score'].min()) / (df['sentiment_score'].max() - df['sentiment_score'].min())

# Frustration Index = 1 - avg of normalized satisfaction signals
df['frustration_index'] = 1 - (0.5 * df['norm_rate'] + 0.5 * df['norm_sentiment'])
```

Content-Based Recommendation System

```python
# Features to include in similarity
feature_cols = [
    'num_dishes_liked',
    'approx_cost(for two people)',
    'rate',
    'type_Buffet', 'type_Cafes', 'type_Delivery', 'type_Desserts',
    'type_Dine-out', 'type_Drinks & nightlife', 'type_Pubs and bars'
]

features = df[feature_cols].copy()

# Normalize features
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# First ensure selected features have no NaNs
features = df[feature_cols].copy()
features = features.fillna(0)  # or use .dropna() if you're okay with removing rows
```

```python
# Normalize
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# Now compute cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(features_scaled)
```

```python
# Helper: Get index from name
def get_index_by_name(name):
    return df[df['name'].str.lower() == name.lower()].index[0]

# Recommend top N similar restaurants
def recommend_restaurants(restaurant_name, top_n=5):
    try:
        idx = get_index_by_name(restaurant_name)
    except IndexError:
        print(f"Restaurant '{restaurant_name}' not found.")
        return

    sim_scores = list(enumerate(similarity_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Exclude itself (index 0)
    sim_scores = sim_scores[1:top_n + 1]

    print(f"\n🍽️ Recommendations similar to '{restaurant_name}':")
    for i, score in sim_scores:
        print(f"➡️ {df.loc[i, 'name']} | Rating: {df.loc[i, 'rate']} | Cost: ₹{df.loc[i, 'approx_cost(for two people)']}")
```

```python
recommend_restaurants("Jalsa", top_n=5)
```

🍽️ Cuisine-Based Clustering by Location

```python
# Fill NaNs just in case
df['cuisines'] = df['cuisines'].fillna('')

# Split cuisines into list
df['cuisine_list'] = df['cuisines'].apply(lambda x: [c.strip() for c in x.split(',')])

# Explode and one-hot encode
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
cuisine_ohe = pd.DataFrame(mlb.fit_transform(df['cuisine_list']), columns=mlb.classes_, index=df.index)

# Add one-hot encoding to original df
df = pd.concat([df, cuisine_ohe], axis=1)
```

```python
cuisine_by_location = df.groupby('location')[mlb.classes_].sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cuisine_scaled = scaler.fit_transform(cuisine_by_location)
```

```python
from sklearn.cluster import KMeans

# Choose number of clusters (try 3–6)
kmeans = KMeans(n_clusters=4, random_state=42)
cuisine_by_location['cluster'] = kmeans.fit_predict(cuisine_scaled)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.countplot(x='cluster', data=cuisine_by_location.reset_index(), palette='tab10')
plt.title('Number of Locations per Cuisine Cluster')
plt.xlabel('Cluster')
plt.ylabel('Number of Locations')
plt.tight_layout()
plt.show()
```

```python
cluster_summary = cuisine_by_location.groupby('cluster').mean().T

# Show top 5 cuisines per cluster
for c in cluster_summary.columns:
    print(f"\n🍽️ Cluster {c} Top Cuisines:")
    print(cluster_summary[c].sort_values(ascending=False).head(5))
```
